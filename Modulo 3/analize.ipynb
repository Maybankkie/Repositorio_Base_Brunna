# Revisão Pandas

```python
 # %pip install pandas

import pandas as pd 

# criar um data frame, trazer os dados para o pandas 
df = pd.read_csv('estoque.csv')

display(df)
```

```python
# ver apenas as primeiras linhas 
print(df.head(2))


print('\n------------------\n')


# ver apenas as colunas do conjunto de dados
for colunas in df.columns:
    print(colunas)
```

```python
#imprimir apenas uma coluna
print(df['modelo'])

print('\n______________________________________\n')

#imprimir varias colunas
print(df[['modelo','valor']].head())
```

```python
# agrupar os dados !!!
# exemplo ver o estoque por cor 
cores = df.groupby('cor')['cor'].count()
print(cores)

print('\n____________________________________\n')

#contar o estoque por modelo
modelo = df.groupby('modelo')['modelo'].count()
print(modelo)
```

```python
# Criar uma coluna nova, ex. faturamento
df['faturamento'] = df['quantidade'] * df['valor']

display(df.head())

print('\n___________________________________________\n')

# qual é o valor total desse estoque
faturamentoTotal = df['faturamento'].sum()
print(f'Valor total em estoque: R$ {faturamentoTotal:.2f} reais')
```

```python
# Filtros
# ex. filtar produtos por um tamanho 
#tamanhoP = df[ df['tamanho'] == 'P' ]
#print(tamanhoP)

#crie um filtro em que o usuario digite o tamanho desejado

tamanho = input('qual o tamanho desejado: ')
filtroTamanho = df[ df['tamanho'] == tamanho ]
print(filtroTamanho)
```
